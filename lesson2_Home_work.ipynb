```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

#https://www.kaggle.com/c/boston-housing/overview

df = pd.read_csv('homework.csv')
df
```

```python
#проверяем на NA
df.info()
#NA отсутствуют
```

```python
#посмотрим на данные с точки зрения статистики
df.describe()
```

```python
#делим выборку на тестовую и тренировочную выборки
x = df[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat']]
y = df[['medv']]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
```

```python
#делаем "стандартизацию" данных и записываем в формат dataFrame
#stdsc = StandardScaler()
#x_train = stdsc.fit_transform(x_train)
#x_test = stdsc.transform(x_test)
#x_train = pd.DataFrame(x_train, columns = x.columns)
#x_test = pd.DataFrame(x_test, columns = x.columns)
```

```python
#Создадим модель линейной регрессии, проведем тренировку на тренировочных данных и вычислим соответствующие коэффициенты для признаков
lr = LinearRegression()
lr.fit(x_train, y_train)
print(lr.coef_, lr.intercept_)
y_predict = lr.predict(x_test)
np.transpose(y_predict)
```

```python
lr_mse = mean_squared_error(y_test, y_predict)
lr_mae = mean_absolute_error(y_test, y_predict)
lr_r2 = r2_score(y_test, y_predict)
print('Mean Square error: ',lr_mse, '\nMean Absolute error: ', lr_mae, '\nR^2: ', lr_r2)
```

```python
#проверка
len_0 = x_test.iloc[[0]].shape[1]
x_arr = list(x_test.iloc[[0]].values[0])
y_arr = list(np.transpose(lr.coef_[0]))
print('Расчетное значение через коэффициенты: ', sum([x_arr[i]*y_arr[i] for i in range(len_0)]) + lr.intercept_[0])
print('Расчетное значение через функцию predict: ', y_predict[0][0])
print('Реальное значение: ', list(y_test.iloc[[0]].values[0])[0])
```

```python
#Создадим модель линейной регрессии SVR
svr = SVR(gamma = 'scale', C=1.0, epsilon=0.2)
svr.fit(np.array(x_train), np.ravel(np.array(y_train)))
```

```python
y_pred_svr = svr.predict(x_test)
y_pred_svr
```

```python
svr_mse = mean_squared_error(y_test, y_pred_svr)
svr_mae = mean_absolute_error(y_test, y_pred_svr)
svr_r2 = r2_score(y_test, y_pred_svr)
print('Mean Square error: ',svr_mse, '\nMean Absolute error: ', svr_mae, '\nR^2: ', svr_r2)
```

```python
#Создадим модель DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(np.array(x_train), np.ravel(np.array(y_train)))
```

```python
y_pred_dtr = dtr.predict(x_test)
y_pred_dtr
```

```python
dtr_mse = mean_squared_error(y_test, y_pred_dtr)
dtr_mae = mean_absolute_error(y_test, y_pred_dtr)
dtr_r2 = r2_score(y_test, y_pred_dtr)
print('Mean Square error: ',dtr_mse, '\nMean Absolute error: ', dtr_mae, '\nR^2: ', dtr_r2)
```

Вывод: все модели показали результат не очень точный, где коэффициент детерминации r2 0.72 был у модели линейной регресии (получается что, лучший метода), то есть модель на 72% кореллирует с исходными данными. хуже результат был у модели дерева решений 0.65 и совсем плохой 0,18. MAE и MSE лучше всех также  у линейной регрессии (менбшее значение, значит меньшая ошибка) и затем у дерева решений у дерева решений. Хуже у SVR
